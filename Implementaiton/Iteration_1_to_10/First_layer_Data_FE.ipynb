```python
import psutil
import GPUtil
import platform
import os
import subprocess

def get_system_info():
    system_info = {
        "System": platform.system(),
        "Node Name": platform.node(),
        "Release": platform.release(),
        "Version": platform.version(),
        "Machine": platform.machine(),
        "Processor": platform.processor(),
        "Architecture": platform.architecture()[0],
    }
    return system_info

def get_ram_info():
    ram = psutil.virtual_memory()
    swap = psutil.swap_memory()
    ram_info = {
        "Total RAM (GB)": round(ram.total / (1024 ** 3), 2),
        "Available RAM (GB)": round(ram.available / (1024 ** 3), 2),
        "Used RAM (GB)": round(ram.used / (1024 ** 3), 2),
        "RAM Usage (%)": ram.percent,
        "Total Swap (GB)": round(swap.total / (1024 ** 3), 2),
        "Used Swap (GB)": round(swap.used / (1024 ** 3), 2),
        "Swap Usage (%)": swap.percent,
    }
    return ram_info

def get_gpu_info():
    gpus = GPUtil.getGPUs()
    gpu_info = []
    for gpu in gpus:
        gpu_info.append({
            "GPU Name": gpu.name,
            "GPU Driver": gpu.driver,
            "Total Memory (GB)": round(gpu.memoryTotal / 1024, 2),
            "Used Memory (GB)": round(gpu.memoryUsed / 1024, 2),
            "Free Memory (GB)": round(gpu.memoryFree / 1024, 2),
            "GPU Load (%)": gpu.load * 100,
            "Temperature (°C)": gpu.temperature,
        })
    return gpu_info

def get_storage_info():
    partitions = psutil.disk_partitions()
    storage_info = {}
    for partition in partitions:
        try:
            usage = psutil.disk_usage(partition.mountpoint)
            storage_info[partition.device] = {
                "Mount Point": partition.mountpoint,
                "File System": partition.fstype,
                "Total Space (GB)": round(usage.total / (1024 ** 3), 2),
                "Used Space (GB)": round(usage.used / (1024 ** 3), 2),
                "Free Space (GB)": round(usage.free / (1024 ** 3), 2),
                "Usage (%)": usage.percent,
            }
        except PermissionError:
            continue
    return storage_info

def main():
    output = []
    output.append("System Information:")
    for key, value in get_system_info().items():
        output.append(f"{key}: {value}")
    
    output.append("\nRAM Information:")
    for key, value in get_ram_info().items():
        output.append(f"{key}: {value}")
    
    output.append("\nGPU Information:")
    for gpu in get_gpu_info():
        for key, value in gpu.items():
            output.append(f"{key}: {value}")
        output.append("")
    
    output.append("Storage Information:")
    for device, info in get_storage_info().items():
        output.append(f"Device: {device}")
        for key, value in info.items():
            output.append(f"  {key}: {value}")
        output.append("")
    
    with open("system_info.txt", "w") as file:
        file.write("\n".join(output))

if __name__ == "__main__":
    main()
```

To maintain a scalable and efficient IDS, we will prioritise **numeric feature selection** and **dimensionality reduction techniques** instead of full categorical encoding.

### Feature selection for Object columns

```python
# Extracting object (categorical) columns
cat_features = df.select_dtypes(include=['object']).columns.tolist()
print("Categorical Features:", cat_features)

# Saving extracted categorical features for reference
df[cat_features].to_csv("categorical_features.csv", index=False)
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()
le = LabelEncoder()

# Encode categorical features
for col in cat_features:
    df_encoded[col] = le.fit_transform(df[col])

# Save encoded dataset
df_encoded.to_csv("encoded_dataset.csv", index=False)
print("Categorical Features Encoded Successfully.")
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()
le = LabelEncoder()

# Converting all categorical columns to string format to prevent mixed types
for col in cat_features:
    df_encoded[col] = df_encoded[col].astype(str)  # Converting to string before encoding
    df_encoded[col] = le.fit_transform(df_encoded[col])  # Applying Label Encoding

# Save encoded dataset
df_encoded.to_csv("encoded_dataset.csv", index=False)
print("Categorical Features Encoded Successfully.")
```

```python
from sklearn.feature_selection import mutual_info_classif

# Computing MI score for categorical features against 'Attack_label'
mi_scores = mutual_info_classif(df_encoded[cat_features], df_encoded["Attack_label"])

# Storing results in a DataFrame
mi_results = pd.DataFrame({'Feature': cat_features, 'MI Score': mi_scores})
mi_results = mi_results.sort_values(by="MI Score", ascending=False)

# Save MI scores
mi_results.to_csv("categorical_feature_importance_mi.csv", index=False)
print("Mutual Information Score Computed and Saved.")
```

```python
from scipy.stats import chi2_contingency

chi2_results = {}

# Computing Chi-Square for each categorical feature
for col in cat_features:
    contingency_table = pd.crosstab(df_encoded[col], df_encoded["Attack_label"])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    chi2_results[col] = p  # Storing p-value

# Converting to DataFrame
chi2_df = pd.DataFrame(list(chi2_results.items()), columns=['Feature', 'P-Value'])
chi2_df = chi2_df.sort_values(by="P-Value")

# Saving Chi-Square results
chi2_df.to_csv("categorical_feature_importance_chi2.csv", index=False)
print("Chi-Square Test Completed. Results Saved.")
```

```python
# Selecting features based on MI and Chi-Square significance
significant_features = mi_results[mi_results['MI Score'] > 0.01]['Feature'].tolist()
significant_features += chi2_df[chi2_df['P-Value'] < 0.05]['Feature'].tolist()
significant_features = list(set(significant_features))  # Remove duplicates

print("Selected Features for IDS:", significant_features)

# Saving selected features for final dataset
df_selected = df_encoded[significant_features + ["Attack_label"]]
df_selected.to_csv("final_selected_features.csv", index=False)
print("Final Feature Set Saved.")
```
