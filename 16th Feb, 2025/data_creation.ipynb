```python
import pandas as pd
import numpy as np

def sample_and_combine_labels(df, sample_size=5000):
    """
    Sample data for each label and combine into a single DataFrame.
    
    Args:
        df: Input DataFrame
        sample_size: Number of samples to take from each label
    
    Returns:
        Combined DataFrame with sampled data from all labels
    """
    # Initialize empty list to store sampled data
    sampled_data_list = []
    
    # Get unique labels
    labels = df['label'].unique()
    
    # Dictionary to store sample counts
    sample_counts = {}
    
    # Process each label
    for label in labels:
        # Get data for current label
        label_data = df[df['label'] == label]
        
        # Determine sample size for this label
        actual_sample_size = min(sample_size, len(label_data))
        
        # Sample the data
        if actual_sample_size == len(label_data):
            sampled_data = label_data  # Take all data if less than sample_size
        else:
            sampled_data = label_data.sample(n=actual_sample_size, random_state=42)
        
        # Append to list
        sampled_data_list.append(sampled_data)
        
        # Store count for reporting
        sample_counts[label] = actual_sample_size
        
        print(f"Processed {label}: {actual_sample_size} samples")
    
    # Combine all sampled data
    combined_df = pd.concat(sampled_data_list, ignore_index=True)
    
    # Print summary
    print("\nSampling Summary:")
    print("-" * 50)
    for label, count in sample_counts.items():
        print(f"{label}: {count} samples")
    print("-" * 50)
    print(f"Total samples in combined dataset: {len(combined_df)}")
    
    return combined_df

# Main execution
try:
    # Sample and combine data
    sampled_df = sample_and_combine_labels(df, sample_size=5000)
    
    # Save combined dataset
    output_file = 'sampled_combined_dataset.csv'
    sampled_df.to_csv(output_file, index=False)
    print(f"\nSaved combined dataset to {output_file}")
    
    # Display distribution in final dataset
    print("\nDistribution in final dataset:")
    print(sampled_df['label'].value_counts())
    
except Exception as e:
    print(f"An error occurred: {str(e)}")
```

```python
import pandas as pd
import numpy as np

def convert_datatypes(df):
    # Creating a copy to avoid modifying the original DataFrame
    df_converted = df.copy()
    
    # List of columns that should be float
    float_columns = [
        'flow_duration', 'Header_Length', 'Duration', 'Rate', 
        'Srate', 'Drate', 'fin_flag_number', 'syn_flag_number',
        'rst_flag_number', 'psh_flag_number', 'ack_flag_number',
        'ece_flag_number', 'cwr_flag_number', 'ack_count',
        'syn_count', 'fin_count', 'urg_count', 'rst_count',
        'Tot sum', 'Min', 'Max', 'AVG', 'Std', 'Tot size',
        'IAT', 'Number', 'Magnitue', 'Radius', 'Covariance',
        'Variance', 'Weight',
        'HTTP', 'HTTPS', 'DNS', 'Telnet', 'SMTP', 'SSH', 
        'IRC', 'TCP', 'UDP', 'DHCP', 'ARP', 'ICMP', 'IPv', 'LLC'
    ]
    
    # Convert numeric columns
    for col in float_columns:
        try:
            # First, handle any string-based numbers
            df_converted[col] = pd.to_numeric(df_converted[col], errors='coerce')
        except Exception as e:
            print(f"Warning: Could not convert {col}: {str(e)}")
    
    # Convert Protocol Type to categorical
    df_converted['Protocol Type'] = df_converted['Protocol Type'].astype('category')
    
    # Keep label as is (categorical)
    df_converted['label'] = df_converted['label'].astype('category')
    
    # Print conversion summary
    print("\nDatatype Conversion Summary:")
    print("-" * 50)
    print("\nNew datatypes:")
    print(df_converted.dtypes)
    
    # Print memory usage comparison
    original_memory = df.memory_usage(deep=True).sum() / 1024**2
    converted_memory = df_converted.memory_usage(deep=True).sum() / 1024**2
    print(f"\nMemory usage reduced from {original_memory:.2f} MB to {converted_memory:.2f} MB")
    print(f"Memory savings: {(original_memory - converted_memory):.2f} MB ({((original_memory - converted_memory)/original_memory)*100:.1f}%)")
    
    return df_converted

# Main execution
try:
    # Convert datatypes
    df_new = convert_datatypes(df)
    
    # Verify successful conversion
    print("\nVerification of numeric columns:")
    print(df_new.select_dtypes(include=['float64', 'int64']).columns.tolist())
    
    # Save converted dataset
    output_file = 'converted_dataset.csv'
    df_new.to_csv(output_file, index=False)
    print(f"\nSaved converted dataset to {output_file}")
    
except Exception as e:
    print(f"An error occurred: {str(e)}")
```
